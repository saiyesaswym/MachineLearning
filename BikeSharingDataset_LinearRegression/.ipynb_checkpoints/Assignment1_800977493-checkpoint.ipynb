### Super Classs Definition

```python
import numpy as np 
from abc import ABC, abstractmethod

# Super class for machine learning models 

class BaseModel(ABC):
    """ Super class for ITCS Machine Learning Class"""
    
    @abstractmethod
    def train(self, X, T):
        pass

    @abstractmethod
    def use(self, X):
        pass

    
class LinearModel(BaseModel):
    """
        Abstract class for a linear model 
        
        Attributes
        ==========
        w       ndarray
                weight vector/matrix
    """

    def __init__(self):
        """
            weight vector w is initialized as None
        """
        self.w = None

    def _check_matrix(self, mat, name):
        if len(mat.shape) != 2:
            raise ValueError(''.join(["Wrong matrix ", name]))
        
    # add a basis
    def add_ones(self, X):
        """
            add a column basis to X input matrix
        """
        self._check_matrix(X, 'X')
        return np.hstack((np.ones((X.shape[0], 1)), X))

    ####################################################
    #### abstract funcitons ############################
    @abstractmethod
    def train(self, X, T):
        """
            train linear model
            
            parameters
            -----------
            X     2d array
                  input data
            T     2d array
                  target labels
        """        
        pass
    
    @abstractmethod
    def use(self, X):
        """
            apply the learned model to input X
            
            parameters
            ----------
            X     2d array
                  input data
            
        """        
        pass 
```

### Least Squares

```python
# Linear Regression Class for least squares
class LinearRegress(LinearModel): 
    """ 
        LinearRegress class 
        
        attributes
        ===========
        w    nd.array  (column vector/matrix)
             weights
    """
    def __init__(self):
        LinearModel.__init__(self)
        
    # train lease-squares model
    def train(self, X, T):
        X1 = self.add_ones(X)
        self.w = np.linalg.inv(X1.T @ X1) @ X1.T @ T 
        return self.w
    
    # apply the learned model to data X
    def use(self, X):
        X1 = self.add_ones(X)
        y = X1 @ self.w
        return y
```

### Least Squares Model:
LinearRegress() contains two important methods: train() and use()

train() function takes two parameters X - input parameters and T - target variable. The input variables are stacked with a column of 1's using the method in the Base Class 'add_ones()'. In the next step, parameters 'w' is calculated using the 'np.linalg.inv()' method. The parameter 'w' is returned by the function. 

use() function takes only input parameters X. Using the same principle, X1 is calculates using 'add_ones()' method. Then the target variable is predicted 'y', by calculating the DOT product of X1 and w. 

### Least Mean Squares

```python
import collections # for checking iterable instance

# LMS class 
class LMS(LinearModel):
    """
        Lease Mean Squares. online learning algorithm
    
        attributes
        ==========
        w        nd.array
                 weight matrix
        alpha    float
                 learning rate
    """
    def __init__(self, alpha):
        LinearModel.__init__(self)
        self.alpha = alpha  
 
    # batch training by using train_step function
    def train(self, X, T):
        N = X.shape[0]
        X1 = self.add_ones(X)
        self.w = np.zeros(X1.shape[1])
        #For every value in the X and T, train_step function is called.
        for x,t in zip(X,T):
            self.w = self.train_step(x,t)
        return self.w
            
    # train LMS model one step 
    # here the x is 1d vector
    def train_step(self, x, t):
        x1 = np.concatenate([[1],x])
        #If w is None, we initiate it to a numpy array of zeroes.This will happen for the first time.
        if self.w is None:
            self.w = np.zeros(x1.shape[0])
        #The target variable is predicted by multiplying w and x1
        y = self.w @ x1
        #Parameter 'w' is updated by subtracting the value 
        self.w -= self.alpha * (y - t) * x1
        return self.w
    
    # apply the current model to data X
    def use(self, X):
        X1 = self.add_ones(X)
        Y = X1 @ self.w
        Y = np.reshape(Y, (-1, 1))
        return Y
```

### Least Mean Squares Model:
LMS() class contains three important functions: train_step(), train() and use()

train_step() - In this case, the input would be single dimension input and target variables - x and t. In the least mean squares model, the parameters 'w' are updated everytime. 
So initially, if the 'w' is None or not initiated, we initiate it to a numpy array of zeroes. 
Then the predicted values are caluclated based on the initiated 'w' values. 
Finally the parameter value is updated by subtracting it from the value of (alpha . (y - t) . x1). 

train() - This function includes calculating the final 'w' which involves all the intermediate steps. Hence after w is initiated, for every value in X and T train_step() is called.

use() function takes only input parameters X. Using the same principle, X1 is calculates using 'add_ones()' method. Then the target variable is predicted 'Y', by calculating the DOT product of X1 and w.
